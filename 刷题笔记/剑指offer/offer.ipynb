```python
#重建二叉树
# -*- coding:utf-8 -*-
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
class Solution:
    # 返回构造的TreeNode根节点
    def reConstructBinaryTree(self, pre, tin):
        # write code here
        if(len(pre)==0):
            return None
        elif(len(pre)==1):
            return TreeNode(pre[0])
        else:
            root = TreeNode(pre[0])
            pos = tin.index(pre[0])
            root.left = self.reConstructBinaryTree(pre[1:pos+1],tin[:pos])
            root.right = self.reConstructBinaryTree(pre[pos+1:],tin[pos+1:])
        return root
```

```python
##二叉树的镜像
class Solution:
    # 返回镜像树的根节点
    def Mirror(self, root):
        # write code here
        if not root:
            return None
        if(root.left == None and root.right == None):
            return root
        if root.left or root.right:
            root.left,root.right = root.right,root.left
            self.Mirror(root.left)
            self.Mirror(root.right)
        return root
```

```python
#从上往下打印二叉树
# -*- coding:utf-8 -*-
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
class Solution:
    # 返回从上到下每个节点值列表，例：[1,2,3]
    def PrintFromTopToBottom(self, root):
        # write code here
        if not root:
            return []
        result = []
        arr = [root]
        while len(arr):
            cur = arr.pop(0)
            result.append(cur.val)
            if(cur.left):
                arr.append(cur.left)
            if(cur.right):
                arr.append(cur.right)
        return result
##用队列,放进去一个就把左右两个插在后面.
```

```python
#二叉树的下一个结点
class Solution:
    def GetNext(self, pNode):
        # write code here
        if not pNode:
            return None
        if pNode.right:
            result = pNode.right
            while(result.left):
                result = result.left
            return result
        else:
            if(pNode.next == None):#根节点
                return pNode.right
            elif(pNode.next.left == pNode):#左节点
                return pNode.next
            elif(pNode.next.right == pNode):#右节点
                while(pNode.next):
                    if pNode.next.right == pNode:
                        pNode = pNode.next
                    else:break
                if pNode.next:
                    if pNode.next.left:
                        result = pNode.next
                        return result
                else:return None
```

```python
#58.对称的二叉树
class Solution:
    def isSymmetrical(self, pRoot):
        # write code here
        def is_same(p1,p2):
            if not p1 and not p2:
                return True
            if (p1 and p2) and p1.val==p2.val:
                return is_same(p1.left,p2.right) and is_same(p1.right,p2.left)##对比左右
            return False
        
        if not pRoot:
            return True
        if (pRoot.left and not pRoot.right) or (not pRoot.left and pRoot.right):
            return False
        return is_same(pRoot.left,pRoot.right)
```

```python
# 序列化二叉树
class Solution:
    flag = -1
    def Serialize(self, root):
        s = ""
        s = self.recursionSerialize(root, s)
        return s
    
    def recursionSerialize(self, root, s):#前序遍历
        if (root is None):
            s =s+'$,'#没东西就用$占位
            return s
        s += str(root.val) + ','
        s = self.recursionSerialize(root.left, s)
        s = self.recursionSerialize(root.right, s)
        return s
    
    def Deserialize(self, s):##重构二叉树
        self.flag += 1
        l = s.split(',')
        if (self.flag >= len(s)):
            return None
        root = None
        if (l[self.flag] != '$'):
            root = TreeNode(int(l[self.flag]))
            root.left = self.Deserialize(s)
            root.right = self.Deserialize(s)
        return root
```

```python
#二叉搜索树的后序遍历序列
# -*- coding:utf-8 -*-
#判断左子树和右子树是不是分别比自己小或者大
class Solution:
    def VerifySquenceOfBST(self, sequence):
        # write code here
        len_s = len(sequence)
        if not len_s:
            return False#不存在
        if len_s==1:
            return True#只有一个结点
        root = sequence[-1]
        pos = 0
        
        while(sequence[pos]<root):
            pos=pos+1
            
        for i in range(pos,len_s-1):
            if sequence[i]<root:
                return False
            
        left = True
        right = True
        if(pos>0):
            left=self.VerifySquenceOfBST(sequence[0:pos])
        if(len_s-1>pos):
            right=self.VerifySquenceOfBST(sequence[pos:len_s-1])
        return left and right
        
```

```python
#二叉搜索树与双向链表
#解法1:调整指针方向
class Solution:
    def Convert(self, pRootOfTree):
        if not pRootOfTree:
            return pRootOfTree
        if not pRootOfTree.left and not pRootOfTree.right:
            return pRootOfTree
        # 处理左子树
        self.Convert(pRootOfTree.left)
        left=pRootOfTree.left
 
        # 连接根与左子树最大结点
        if left:
            while(left.right):
                left=left.right
            pRootOfTree.left,left.right=left,pRootOfTree
        #####

        # 处理右子树
        self.Convert(pRootOfTree.right)
        right=pRootOfTree.right
 
        # 连接根与右子树最小结点
        if right:
            while(right.left):
                right=right.left
            pRootOfTree.right,right.left=right,pRootOfTree
        #####
        while(pRootOfTree.left):
            pRootOfTree=pRootOfTree.left
        return pRootOfTree

#解法2:用一个数组保存指针,最后统一管理(推荐推荐)
class Solution:
    def NodeList(self, pRootOfTree):
        if not pRootOfTree:
            return []
        return self.NodeList(pRootOfTree.left) + [pRootOfTree] + self.NodeList(pRootOfTree.right)
    def Convert(self, pRootOfTree):
        # write code here
        res = self.NodeList(pRootOfTree)
        if len(res) == 0:
            return None
        if len(res) == 1:
            return pRootOfTree
        res[0].left = None
        res[0].right = res[1]
        res[-1].left = res[-2]
        res[-1].right = None
        for i in range(1,len(res)-1):
            res[i].left = res[i-1]
            res[i].right = res[i+1]
        return res[0]
```

```python
#包含min函数的栈
class Solution:
    def __init__(self):
        self.stack = []
        self.min_stack = []#用于排序.
    def push(self, node):
        # write code here
        self.stack.append(node)
        if not self.min_stack or node <= self.min_stack[-1]:#不存在的时候就放入栈中
            self.min_stack.append(node)
    def pop(self):
        # write code here
        if self.stack[-1] == self.min_stack[-1]:
            self.min_stack.pop()
        self.stack.pop()
    def top(self):
        # write code here
        return self.stack[-1]
    def min(self):
        # write code here
        return self.min_stack[-1]
```

```python
#数组中出现超过一半的数字
class Solution:
    def MoreThanHalfNum_Solution(self, numbers):
        # write code here
        if not numbers:
            return 0
        if len(numbers)==1:
            return numbers[0]
        numbers = sorted(numbers)
        lens = int(len(numbers)/2)
        for i in range(0,lens):
            if numbers[i] == numbers[i+lens]:
                return numbers[i]
        return 0
```

# 快速排序

```python
def quicksort(array,l,r):
    if l < r:
        key = partition(array,l,r)
        quicksort(array,l,key-1)
        quicksort(array,key+1,r)
def partition(array,l,r):
    key = array[r]
    i = l-1
    for j in range(l,r):
        if array[j]<key:
            i+=1
            array[j],array[i] = array[i],array[j]
        array[i+1],array[r] = array[r],array[i+1]
    return i+1

a = [1,2,2,34,2]
quicksort(a,0,len(a)-1)
a
```

```python
def prime(number,l,r):
    if(number == 0 or number == 1):
        return False
    elif(l>=r):
        return False
    else:
        if(number%l!=0):
            return prime(number,l+1,int(number/l))
        else:
            return True
f = open('prime.txt','w')
for i in range(2,1000):
    if not prime(i,2,i):
        f.write(str(i)+'\n')
        print(i)

f.close()
```

```python
#栈的压入弹出序列
class Solution:
    def IsPopOrder(self, pushV, popV):
        # write code here
        stack = []#辅助
        stack.append(pushV.pop(0))
        while(True):
            if(stack ==[] and pushV==[] and popV==[]):
                return True
            if(stack[-1]!=popV[0]):
                if(pushV==[]):
                    return False
                stack.append(pushV.pop(0))
            else:
                popV.pop(0)
                stack.pop(-1)
```

```python
#数组中的逆序对
a  = [1,2,3,4,5,6,7,0]
class Solution:
    def InversePairs(self, data):
        # write code here
        if not data:
            return 0
        res = 0
        s = sorted(data,reverse=True)
        ss = len(s)
        for i in data:
            a = s.index(i)
            
            while(a<ss-1):
                if(s[a+1]==s[a]):
                    a+=1
                else:
                    break
            res+=(ss - a -1)
            s.remove(i)
            ss-=1
        return res%1000000007
h =Solution()
h.InversePairs(a)
```

```python
#孩子们的游戏
class Solution:
    def LastRemaining_Solution(self, n, m):
        # write code here
        if not m or not n:
            return -1
        res = range(n)
        i = 0
        while len(res)>1:
            i = (m+i-1)%len(res)
            res.pop(i)
        return res[0]
```

```python
#把字符串转换为整数
# -*- coding:utf-8 -*-
class Solution:
    def StrToInt(self, s):
        # write code here
        if not s:
            return 0
        flag = True
        s = list(s)
        a=s[0]
        if(a=='+'):
            flag = True
            s.pop(0)
        elif(a=='-'):
            flag = False
            s.pop(0)
        s = s[::-1]
        num=0
        res=0
        while(s!=[]):
            a=s.pop(0)
            if(a>='0' and a<='9'):
                res+= int(a)*(10**num)
                num+=1
            else:
                return 0
        if((res>2147483648 and flag==False) or(res>2147483647 and flag==True)):
            return 0
        return res if flag ==True else -res
s = Solution()
s.StrToInt("-1231")
```

```python
#数组中重复的数字
# -*- coding:utf-8 -*-s
import collections
class Solution:
    # 这里要特别注意~找到任意重复的一个值并赋值到duplication[0]
    # 函数返回True/False
    def duplicate(self, numbers, duplication):
        # write code here
        c = collections.Counter(numbers)
        for k,v in c.items():
            if v>1:
                duplication[0]=k
                return True
        return False
        
```

```python
#机器人
class Solution:
    def movingCount(self, threshold, rows, cols):
        # write code here
        board = [[0 for _ in range(cols)] for _ in range(rows)]
        def block(r,c):
            s = sum(map(int,str(r)+str(c)))
            return s>threshold
        class Context:
            acc = 0
        def traverse(r,c):
            if not (0<=r<rows and 0<=c<cols): return
            if board[r][c]!=0: return
            if board[r][c]==-1 or block(r,c):
                board[r][c]=-1
                return
            board[r][c] = 1
            Context.acc+=1
            traverse(r+1,c)
            traverse(r-1,c)
            traverse(r,c+1)
            traverse(r,c-1)
        traverse(0,0)
        return Context.acc
```

```python
#正则表达式
class Solution:
    # s, pattern都是字符串
    def match(self, s, pattern):
        # write code here
        if (len(s) == 0 and len(pattern) == 0):
            return True
        if (len(s) > 0 and len(pattern) == 0):
            return False
        if (len(pattern) > 1 and pattern[1] == '*'):
            if (len(s) > 0 and (s[0] == pattern[0] or pattern[0] == '.')):
                return (self.match(s, pattern[2:]) or self.match(s[1:], pattern[2:]) or self.match(s[1:], pattern))
            else:
                return self.match(s, pattern[2:])
        if (len(s) > 0 and (pattern[0] == '.' or pattern[0] == s[0])):
            return self.match(s[1:], pattern[1:])
        return False
```

```python
#滑动窗口的最大值
class Solution:
    def maxInWindows(self, num, size):
        if (num is None or len(num) == 0 or size == 0 or len(num) < size):
            return []
        maxList = []
        for i in range(len(num) - size + 1):
            maxList.append(max(num[i:i+size]))
        return maxList
a = [1,213,2,123,12,3]
s = Solution()
s.maxInWindows(a,2)
```

```python
#矩阵中的路径
class Solution:
    def hasPath(self, matrix, rows, cols, path):
        # write code here
        if path == "":
            return True
        if cols <= 0 or rows <= 0:
            return False
         
        matrix = [matrix[r*cols:(r+1)*cols] for r in xrange(0, rows)]
        print matrix
        arrive = [[0 for x in xrange(0, cols)] for y in xrange(0, rows)]
         
         
        def dfs(mat, arr, pat, pos):
            #search path in mat from arr
             
            if pat == "":
                return True
            x ,y = pos
            if matrix[y][x] != pat[0]:
                return False
 
            if len(pat) == 1:
                return True
 
            arr[y][x] = 1
            findflag = False
            if x > 0 and arr[y][x-1] != 1:
                findflag |= dfs(mat, arr, pat[1:], (x-1,y))
            if y > 0 and arr[y-1][x] != 1:
                findflag |= dfs(mat, arr, pat[1:], (x,y-1))
            if x < cols - 1 and arr[y][x+1] != 1:
                findflag |= dfs(mat, arr, pat[1:], (x+1,y))
            if y < rows - 1 and arr[y+1][x] != 1:
                findflag |= dfs(mat, arr, pat[1:], (x,y+1))
            arr[y][x] = 0
            # print "Searching ", pos, pat, findflag
            return findflag
         
        for x in xrange(0, cols):
            for y in xrange(0,rows):
                if dfs(matrix, arrive, path, (x, y)) == True:
                    return True
                 
        return False
```
