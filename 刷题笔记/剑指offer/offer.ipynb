```python
#重建二叉树
# -*- coding:utf-8 -*-
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
class Solution:
    # 返回构造的TreeNode根节点
    def reConstructBinaryTree(self, pre, tin):
        # write code here
        if(len(pre)==0):
            return None
        elif(len(pre)==1):
            return TreeNode(pre[0])
        else:
            root = TreeNode(pre[0])
            pos = tin.index(pre[0])
            root.left = self.reConstructBinaryTree(pre[1:pos+1],tin[:pos])
            root.right = self.reConstructBinaryTree(pre[pos+1:],tin[pos+1:])
        return root
```

```python
class Solution:
    def HasSubtree(self, pRoot1, pRoot2):
        # write code here
        if(pRoot1 == None or pRoot2 == None):
            return False
        if(self.sameroot(pRoot1,pRoot2)):##是子结构吗
            return True
        elif(self.HasSubtree(pRoot1.left,pRoot2)):##找左子树
            return True
        elif(self.HasSubtree(pRoot1.right,pRoot2)):##找右子树
            return True
        else:
            return False
    def sameroot(self,p1,p2):
        if not p2:
            return True
        if not p1 or p1.val!=p2.val:
            return False
        else:
            return self.sameroot(p1.left,p2.left) and self.sameroot(p1.right,p2.right)
```

```python
##二叉树的镜像
class Solution:
    # 返回镜像树的根节点
    def Mirror(self, root):
        # write code here
        if not root:
            return None
        if(root.left == None and root.right == None):
            return root
        if root.left or root.right:
            root.left,root.right = root.right,root.left
            self.Mirror(root.left)
            self.Mirror(root.right)
        return root
```

```python
#从上往下打印二叉树
# -*- coding:utf-8 -*-
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
class Solution:
    # 返回从上到下每个节点值列表，例：[1,2,3]
    def PrintFromTopToBottom(self, root):
        # write code here
        if not root:
            return []
        result = []
        arr = [root]
        while len(arr):
            cur = arr.pop(0)
            result.append(cur.val)
            if(cur.left):
                arr.append(cur.left)
            if(cur.right):
                arr.append(cur.right)
        return result
##用队列,放进去一个就把左右两个插在后面.
```

```python
#二叉树的深度
class Solution:
    def TreeDepth(self, pRoot):
        # write code here
        if not pRoot:
            return 0
        left = self.TreeDepth(pRoot.left)
        right = self.TreeDepth(pRoot.right)
        return left+1 if left>right else right+1
```

```python
#二叉树的下一个结点
class Solution:
    def GetNext(self, pNode):
        # write code here
        if not pNode:
            return None
        if pNode.right:
            result = pNode.right
            while(result.left):
                result = result.left
            return result
        else:
            if(pNode.next == None):#根节点
                return pNode.right
            elif(pNode.next.left == pNode):#左节点
                return pNode.next
            elif(pNode.next.right == pNode):#右节点
                while(pNode.next):
                    if pNode.next.right == pNode:
                        pNode = pNode.next
                    else:break
                if pNode.next:
                    if pNode.next.left:
                        result = pNode.next
                        return result
                else:return None
```

```python
#58.对称的二叉树
class Solution:
    def isSymmetrical(self, pRoot):
        # write code here
        def is_same(p1,p2):
            if not p1 and not p2:
                return True
            if (p1 and p2) and p1.val==p2.val:
                return is_same(p1.left,p2.right) and is_same(p1.right,p2.left)##对比左右
            return False
        
        if not pRoot:
            return True
        if (pRoot.left and not pRoot.right) or (not pRoot.left and pRoot.right):
            return False
        return is_same(pRoot.left,pRoot.right)
```

```python
# 序列化二叉树
class Solution:
    flag = -1
    def Serialize(self, root):
        s = ""
        s = self.recursionSerialize(root, s)
        return s
    
    def recursionSerialize(self, root, s):#前序遍历
        if (root is None):
            s =s+'$,'#没东西就用$占位
            return s
        s += str(root.val) + ','
        s = self.recursionSerialize(root.left, s)
        s = self.recursionSerialize(root.right, s)
        return s
    
    def Deserialize(self, s):##重构二叉树
        self.flag += 1
        l = s.split(',')
        if (self.flag >= len(s)):
            return None
        root = None
        if (l[self.flag] != '$'):
            root = TreeNode(int(l[self.flag]))
            root.left = self.Deserialize(s)
            root.right = self.Deserialize(s)
        return root
```

```python
#二叉搜索树的后序遍历序列
# -*- coding:utf-8 -*-
#判断左子树和右子树是不是分别比自己小或者大
class Solution:
    def VerifySquenceOfBST(self, sequence):
        # write code here
        len_s = len(sequence)
        if not len_s:
            return False#不存在
        if len_s==1:
            return True#只有一个结点
        root = sequence[-1]
        pos = 0
        
        while(sequence[pos]<root):
            pos=pos+1
            
        for i in range(pos,len_s-1):
            if sequence[i]<root:
                return False
            
        left = True
        right = True
        if(pos>0):
            left=self.VerifySquenceOfBST(sequence[0:pos])
        if(len_s-1>pos):
            right=self.VerifySquenceOfBST(sequence[pos:len_s-1])
        return left and right
        
```

```python
#二叉搜索树与双向链表
#解法1:调整指针方向
class Solution:
    def Convert(self, pRootOfTree):
        if not pRootOfTree:
            return pRootOfTree
        if not pRootOfTree.left and not pRootOfTree.right:
            return pRootOfTree
        # 处理左子树
        self.Convert(pRootOfTree.left)
        left=pRootOfTree.left
 
        # 连接根与左子树最大结点
        if left:
            while(left.right):
                left=left.right
            pRootOfTree.left,left.right=left,pRootOfTree
        #####

        # 处理右子树
        self.Convert(pRootOfTree.right)
        right=pRootOfTree.right
 
        # 连接根与右子树最小结点
        if right:
            while(right.left):
                right=right.left
            pRootOfTree.right,right.left=right,pRootOfTree
        #####
        while(pRootOfTree.left):
            pRootOfTree=pRootOfTree.left
        return pRootOfTree

#解法2:用一个数组保存指针,最后统一管理(推荐推荐)
class Solution:
    def NodeList(self, pRootOfTree):
        if not pRootOfTree:
            return []
        return self.NodeList(pRootOfTree.left) + [pRootOfTree] + self.NodeList(pRootOfTree.right)
    def Convert(self, pRootOfTree):
        # write code here
        res = self.NodeList(pRootOfTree)
        if len(res) == 0:
            return None
        if len(res) == 1:
            return pRootOfTree
        res[0].left = None
        res[0].right = res[1]
        res[-1].left = res[-2]
        res[-1].right = None
        for i in range(1,len(res)-1):
            res[i].left = res[i-1]
            res[i].right = res[i+1]
        return res[0]
```

```python
#包含min函数的栈
class Solution:
    def __init__(self):
        self.stack = []
        self.min_stack = []#用于排序.
    def push(self, node):
        # write code here
        self.stack.append(node)
        if not self.min_stack or node <= self.min_stack[-1]:#不存在的时候就放入栈中
            self.min_stack.append(node)
    def pop(self):
        # write code here
        if self.stack[-1] == self.min_stack[-1]:
            self.min_stack.pop()
        self.stack.pop()
    def top(self):
        # write code here
        return self.stack[-1]
    def min(self):
        # write code here
        return self.min_stack[-1]
```

```python
#数组中出现超过一半的数字
class Solution:
    def MoreThanHalfNum_Solution(self, numbers):
        # write code here
        if not numbers:
            return 0
        if len(numbers)==1:
            return numbers[0]
        numbers = sorted(numbers)
        lens = int(len(numbers)/2)
        for i in range(0,lens):
            if numbers[i] == numbers[i+lens]:
                return numbers[i]
        return 0
```
